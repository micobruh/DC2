```python
df = pd.read_excel("open-data-table-police-workforce-280721.ods", sheet_name = "Data")
df
```

```python
df.rename(columns = {'As at 31 March...': 'Year', 'Geo code': 'Geo_Code', 'Force Name': 'Force_Name', 
                     'Total (headcount)': 'Deployed_Amount'}, inplace = True)
df_need = df[df_need['Year'] >= 2010][['Year', 'Geo_Code', 'Force_Name', 'Deployed_Amount']]
df_need
```

```python
df_deploy = df_need.groupby(['Year', 'Geo_Code', 'Force_Name']).sum().reset_index()
df_deploy
```
