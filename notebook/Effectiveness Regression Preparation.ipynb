```python
import pandas as pd
import numpy as np
```

```python
df_police = pd.read_csv('deployed_police.csv')
df_crime = pd.read_csv('total_crime.csv')
```

```python
df_police['Force_Name'] = df_police['Force_Name'].replace({'Metropolitan Police': 'Metropolitan', 
                                                           'London, City of': 'City of London', 
                                                           'Avon & Somerset': 'Avon and Somerset'})
df_police['Force_Name'] = df_police['Force_Name'].str.replace(' ', '-').str.lower()
df_police_need = df_police[(df_police['Year'] > 2010) & 
                           (df_police['Year'] < 2021)].drop(columns = ['Unnamed: 0']).reset_index(drop = True)
df_police_need
```

```python
df_crime_need = df_crime[(df_crime['Year'] > 2010) & (df_crime['Year'] < 2021) & 
                           (df_crime['Section'] != 'btp')].groupby(['Year', 'Section']).sum()[['Amount']].reset_index()
df_crime_need = df_crime_need.rename(columns = {'Amount': 'Crime_Amount', 'Section': 'Force_Name'})
df_crime_need
```

```python
df_result = pd.merge(df_police_need, df_crime_need, on = ['Year', 'Force_Name'])
df_result
```

```python
df_result.to_csv('effective_regression.csv')
```
