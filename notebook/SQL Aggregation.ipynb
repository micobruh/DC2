```python
import pandas as pd
import sqlite3 as sql
import numpy as np
from tqdm import tqdm
```

```python
conn = sql.connect('crime.db')
```

```python
# All attributes for street_crime: Crime_ID, Month, Reported_by, Falls_within, Longitude, Latitude, Location, LSOA_code, 
# LSOA_name, Crime_type, Last_outcome_category, Context, Section, Year, Month

# All attributes for outcome_crime: Crime_ID, Month, Reported_by, Falls_within, Longitude, Latitude, Location, LSOA_code, 
# LSOA_name, Outcome_type, Section, Year, Month

# All attributes for outcome_stop: Type, Date, Part_of_a_policing_operation, Policing_operation, Latitude, Longitude,
# Gender, Age_range, Self-defined_ethnicity, Officer-defined_ethnicity, Legislation, Object_of_search, Outcome,
# Outcome_linked_to_object_of_search, Removal_of_more_than_just_outer_clothing, Section, Year, Month, Day
```

```python
query_most_crime = '''
    select Year, Month, Section, Crime_type, Amount from
        (select Year, Month, Section, Crime_type, count(Crime_type) 'Amount'
        from street_crime
        group by Year, Month, Section, Crime_type
        order by Year, Month, Section, Amount DESC)
    group by Year, Month, Section
'''
query_least_crime = '''
    select Year, Month, Section, Crime_type, Amount from
        (select Year, Month, Section, Crime_type, count(Crime_type) 'Amount'
        from street_crime
        group by Year, Month, Section, Crime_type
        order by Year, Month, Section, Amount)
    group by Year, Month, Section
'''
query_total_crime = '''
    select Year, Month, Section, count(Longitude) 'Amount'
    from street_crime
    group by Year, Month, Section
'''
df = pd.read_sql_query(query_total_crime, conn)
df
```

```python
df.to_csv("total_crime.csv", index = False)
```
